```python
import requests
import pandas as pd

# DART API 설정
api_key = '0728a6d81b70b86763fd38daafcc1fccff0beedb'
corp_code = '00164742'  # 현대차 고유번호

# DART API 엔드포인트 (수익성 지표)
url = 'https://opendart.fss.or.kr/api/fnlttSinglIndx.json'

# 2019-2023년 사업보고서 데이터 가져오기
financial_data = []

for year in range(2019, 2024):  # 2019 ~ 2023년 데이터 가져오기
    params = {
        'crtfc_key': api_key,   # DART API 키
        'corp_code': corp_code,  # 현대차 고유번호
        'bsns_year': str(year),  # 각 연도의 수익성 지표
        'reprt_code': '11011',   # 사업보고서
        'idx_cl_code': 'M210000' # 수익성 지표 코드 (예: M210000)
    }

    response = requests.get(url, params=params)
    data = response.json()

    # 응답 상태 확인
    if data.get('status') == '000':  # 성공 상태 코드
        df_financial = pd.DataFrame(data['list'])
        df_financial['stlm_dt'] = pd.to_datetime(df_financial['stlm_dt'])  # 날짜 변환
        df_financial = df_financial[['stlm_dt', 'idx_nm', 'idx_val']]  # 필요한 열만 선택
        df_financial.rename(columns={'stlm_dt': 'date', 'idx_nm': 'indicator', 'idx_val': 'value'}, inplace=True)
        df_financial['year'] = year  # 연도 컬럼 추가
        financial_data.append(df_financial)
    else:
        print(f"{year}년 데이터 오류: {data.get('message')}")

# 수익성 지표 데이터 결합
df_financial_all = pd.concat(financial_data, ignore_index=True)
print(df_financial_all.head())
```
