```python
# 결측값 제거
selected_stock = selected_stock.dropna(subset=['Stock_Returns', 'Market_Returns'])

# 숫자 타입으로 변환
selected_stock['Market_Returns'] = pd.to_numeric(selected_stock['Market_Returns'], errors='coerce')
selected_stock['Stock_Returns'] = pd.to_numeric(selected_stock['Stock_Returns'], errors='coerce')

# 결측값 확인
print(selected_stock[['Market_Returns', 'Stock_Returns']].isnull().sum())

# 공분산 계산
cov_matrix = selected_stock[['Stock_Returns', 'Market_Returns']].cov()
cov_stock_market = cov_matrix.loc['Stock_Returns', 'Market_Returns']

selected_stock
```

```python
import pandas as pd

# 모든 종목명 가져오기
종목명_목록 = pricing_data['ISU_NM'].unique()

# 결과 저장용 리스트
beta_results = []

# 각 종목별로 처리
for 종목명 in 종목명_목록:
    print(f"현재 종목: {종목명}")
    
    # 특정 종목 데이터 필터링
    selected_stock = pricing_data[pricing_data['ISU_NM'] == 종목명].copy()

    # BAS_DD를 기준으로 kospi와 병합
    selected_stock = selected_stock.merge(
        kospi[['BAS_DD', 'Market_Returns']],
        on='BAS_DD',
        how='left'
    )

    # 결측값 처리 (문자열을 숫자로 변환)
    selected_stock['Market_Returns'] = pd.to_numeric(selected_stock['Market_Returns'], errors='coerce')
    selected_stock['Stock_Returns'] = pd.to_numeric(selected_stock['Stock_Returns'], errors='coerce')
    selected_stock.dropna(subset=['Market_Returns', 'Stock_Returns'], inplace=True)

    # 베타 계산
    if len(selected_stock) > 1:  # 데이터가 충분히 있어야 계산 가능
        cov_matrix = selected_stock[['Stock_Returns', 'Market_Returns']].cov()
        cov_stock_market = cov_matrix.loc['Stock_Returns', 'Market_Returns']
        market_variance = selected_stock['Market_Returns'].var()
        beta = cov_stock_market / market_variance
        print(f"{종목명}의 베타 값: {beta}\n")
        
        # 결과를 리스트에 저장
        beta_results.append({'종목명': 종목명, '베타 값': beta})
    else:
        print(f"{종목명}의 데이터가 부족하여 베타 값을 계산할 수 없습니다.\n")
        beta_results.append({'종목명': 종목명, '베타 값': None})

```

```python
# 리스트를 데이터프레임으로 변환
beta_df = pd.DataFrame(beta_results)

# CSV 파일로 저장
beta_df.to_csv('beta_values.csv', index=False, encoding='utf-8-sig')

print("베타 값이 beta_values.csv 파일에 저장되었습니다.")
```
